### Mulitple figures

```python
t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)

figure(1)    ### a way to tell where we need to plot
plot(x)
figure(2)
plot(y)

figure(3)   ### lets plot both of them in the same figure
plot(x)
plot(y)
```

#### Mulitple plots using `subplot`

```python
x = array([1,2,3,2,1])
y = array([1,3,2,3,1])

subplot(2, 1, 1)    ### subplot(rows, columns, active plot)
plot(x)

subplot(2, 1, 2)    ### activating new plot by changing the last value
plot(y)
```

#### Adding lines to a plot

The figure is held, if we keep calling plot it will add the line to the same plot, but if we want to have only one figure open and whenever we call plot it should erase everything and shows only the latest plot

If we need to `hold` function will be useful

```python
figure(1)
plot(x)
plot(y)

figure(2)
hold(False)
plot(x)
plot(y)
```

```python
hold(True)
```

### Calculate Derivative

```python
from numpy import linspace,pi, sin, cos, cumsum
from matplotlib.pyplot import plot, show, subplot, legend, title
close('all')
x = linspace(0, 2*pi, 101)
y = sin(x)

dy = y[1:] - y[:-1]
dx = x[1:] - y[:-1]

dy_dx = dy/dx

centers_x = (x[1:] + x[:-1])/2.0

subplot(1,2,1)
plot(centers_x, dy_dx, 'rx', centers_x, cos(centers_x), 'b-')
title(r"$\rm{Derivative\ of}\ sin(x)$")

```
